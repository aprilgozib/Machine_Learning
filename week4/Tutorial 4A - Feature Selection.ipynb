## Week 4: Feature Selection

Reducing the complexity of the data can be done in two ways, `Feature Selection` and `Feature Transformation`. In this notebook, we will focus on `Feature Selection`. Feature Selection implies choosing a subset of the features from the original X. 

#### When to opt for Feature Selection?

- Many X variables are present but **most** of which **contain little information** and some are very meaningful. 
- The meaningful variables are independent of each other. 

Here, we cover univariate feature selection (filters) using `mutual information` and `chi-sq` as well as wrapper techniques using `Sequential feature selection` available on scikit-learn and mlxtend packages respectively. However, there are more ways to implement feature selection on scikit-learn, refer to <a href=https://scikit-learn.org/stable/modules/feature_selection.html#sequential-feature-selection>documentation</a> for details.

#### Import Packages

```python
import pandas as pd
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import chi2, mutual_info_classif, SelectKBest
```

```python
# import wine dataset that is given by scikit-learn, 'as_frame=True' ensures to return the data as a dataframe
wine = datasets.load_wine(as_frame=True)
print(wine.DESCR)
```

```python
X = wine.data
X.head()
```

```python
X.describe().T
```

Since the range of the features vary significantly (for example, `hue` has a max value of 1.71 whereas `proline` has a max value of 1680), features have to be normalized. Here, we will try the `StandardScaler` provided by scikit-learn to standardize the data.

Data Standardization

```python
y = wine.target
y.head()
```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Refer to https://scikit-learn.org/stable/modules/classes.html#module-sklearn.preprocessing for preprocessing data
scaler = StandardScaler()
X_train_transformed = pd.DataFrame(scaler.fit_transform(X_train.values), index=X_train.index, columns=X_train.columns) 
X_train_transformed.head(2)
```

```python
X_test_transformed = pd.DataFrame(scaler.transform(X_test.values), index=X_test.index, columns=X_test.columns) 
X_test_transformed.head(2)
```

### Part A: Filters (Mutual Information)

Mutual information measures the amount of dependency between two random variables and is based on entropy - a value of 0 indicates variables are completely independent. 

```python
i_scores = mutual_info_classif(X_train_transformed, y_train)

# put the i_scores in a dataframe and sort
features_info_gain = pd.DataFrame(i_scores, index = X_train_transformed.columns, columns =['I-Gain'])
features_info_gain.sort_values(by=['I-Gain'], ascending=False, inplace=True)
features_info_gain
```

```python
# Generate top 3 and bottom 3 dataframes and produce the corresponding X (numpy) arrays
top3_features_train= X_train_transformed[features_info_gain.index[:3]]
bottom3_features_train = X_train_transformed[features_info_gain.index[-3:]]

top3_features_test= X_test_transformed[features_info_gain.index[:3]]
bottom3_features_test = X_test_transformed[features_info_gain.index[-3:]]
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(top3_features_train, y_train)

y_predicted = model.predict(top3_features_test)
print("Accuracy of the model trained on the top 3 features: ", accuracy_score(y_test, y_predicted))
```

```python
model.fit(bottom3_features_train, y_train)

y_predicted = model.predict(bottom3_features_test)
print("Accuracy of the model trained on the bottom 3 features: ", accuracy_score(y_test, y_predicted))
```

### Part B: Wrappers (Sequential Feature Selection)

Install mlxtend using the command `conda install mlxtend` in your virtual environment

#### SequentialFeatureSelector

`SequentialFeatureSelector` is made available in scikit-learn recently. However, `mlxtend` package provides better APIs for visualizations. Hence, we will be using the mlxtend package to perform wrapper methods for feature selection. Refer to the documentation of mlxtend <a href=http://rasbt.github.io/mlxtend/user_guide/feature_selection/SequentialFeatureSelector/#sequential-feature-selector>here</a> for more information.

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
sfs_forward = SFS(knn, 
                  k_features="best", 
                  forward=True, 
                  floating=False,
                  scoring='accuracy',
                  cv=5)

# try the sfs_forward model on the 
sfs_forward = sfs_forward.fit(X_train_transformed, y_train)
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
import matplotlib.pyplot as plt

fig1 = plot_sfs(sfs_forward.get_metric_dict(), ylabel='Accuracy')

plt.ylim([0.75, 1])
plt.title('Sequential Forward Selection (w. StdDev)')
plt.grid()
plt.show()
print(sfs_forward.k_feature_names_)
```

Including 4 features has the highest accuracy score. Therefore, 4 seems like a better choice for the number of features for this dataset. To know which features contribute to the high accuracy, we can create a dataframe of the metrics obtained from the sequential forward selection model as given below.

```python
pd.DataFrame.from_dict(sfs_forward.get_metric_dict()).T
```
